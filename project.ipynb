# Education Levels & Weekly Wages (M/F)

```python
# Name / ID: Jona Sahota | 171451870 | Akhil Chaudhary | 160602550

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns

# Reading file 
data = pd.read_csv('wages.csv')
# Filtering the data to the most recent year (2019) added within the File.
dataFrame = data[(data['YEAR'] == 2019) & (data['Geography'] == 'Canada')]

# Grouping by Education level (12 Different Types purposed in File).
education_group = dataFrame.groupby(['Education level']).mean().reset_index().sort_values('Both Sexes', ascending=True)

# Taking the DataFrame and using it to create a specific format of the df where the
# specified columns will work as indentifiers.c
df_to_plot = pd.melt(education_group, id_vars=['Education level'], value_vars=['Male', 'Female', 'Both Sexes'], var_name= 'Gender', value_name='Wages (Weekly)')

# Setting up Figure size to fit everything accordingly.
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)
# Using a colour palette to make it look aesthetically pleasing
current_palette = sns.hls_palette(n_colors=3, h=0.2, l=0.8, s=0.65)

# Plotting the filtered information itself.
# We used the seaborn package to help us out with our needs.
sns.barplot(x="Education level", y="Wages (Weekly)", hue='Gender', data=df_to_plot, ax=ax, palette=current_palette)
plt.xticks(rotation=90)
plt.tight_layout()
```

# Province's & Weekly Wages (Male / Female)

```python
# Name / ID: Jona Sahota | 171451870 | Akhil Chaudhary | 160602550

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
data = pd.read_csv('wages.csv')

# Filtering by getting all the provinces data from the year of 2019.
data = data[(data['YEAR'] == 2019) & data["Geography"].isin(["Ontario",'Quebec','Newfoundland and Labrador','Prince Edward Island','Nova Scotia','New Brunswick','Manitoba','Saskatchewan','Alberta','British Columbia'])]
education_group = data.groupby(['Geography']).mean().reset_index()

# Grouping by Education level (12 Different Types purposed in File).
df_to_plot = pd.melt(data, id_vars=['Geography'], value_vars=['Both Sexes'], var_name= 'Gender', value_name='Wages (Weekly)')

# Setting up Figure size to fit everything accordingly.
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)

# Using a colour palette to make it look aesthetically pleasing
current_palette = sns.hls_palette(n_colors=1, h=0.5, l=0.5, s=0.65)

# Plotting the filtered information itself.
# We used the seaborn package to help us out with our needs.
sns.lineplot(x="Geography", y="Wages (Weekly)", hue='Gender', data=df_to_plot, ax=ax, palette=current_palette)
plt.xticks(rotation=90)
plt.tight_layout()
```
